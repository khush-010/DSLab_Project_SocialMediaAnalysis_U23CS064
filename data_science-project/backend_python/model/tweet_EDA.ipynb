### Combining

```python
import pandas as pd
df1=pd.read_csv('tweets1.csv',encoding='latin-1')
df2=pd.read_csv('tweets2.csv',encoding='latin-1')
```

```python
df=pd.concat([df1,df2],ignore_index=False)
```

```python
df.head()
```

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df.head()
```

```python
df.to_csv('tweets.csv',index=False)
```
